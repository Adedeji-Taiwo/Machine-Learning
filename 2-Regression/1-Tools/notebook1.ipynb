Linear Regression: Diabetes Progression Against BMI

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
X, y = datasets.load_diabetes(return_X_y=True) # X is the 10 column features while y is the single column target or label
print(X.shape)
print(X[0])
```

```python
X = X[:, 2]
X = X.reshape((-1, 1))
print(X)
```

```python
#split both the data (X) and the target (y) into test and training sets 70%-30%
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
#Train model using linear regrression using model fit
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
#create a prediction using test data to draw the line between data groups
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
